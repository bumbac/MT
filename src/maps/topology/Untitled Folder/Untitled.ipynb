```python
import os
import numpy as np
```

```python
def mirror(filename):
    with open(filename, "r") as f:
        lines = f.readlines()
        width = len(lines[0])
        for i in range(100):
            if lines[i][0] == '#':
                continue
            height = i
            break
        goals = []
        for i in range(height+1, len(lines)):
            goals.append(lines[i].rstrip())
        print(goals)

    with open(filename, "w") as f:
        for y in range(height):
            new_line = ""
            for x in range(width):
                new_line += str(lines[height - y - 1][x])
            print(new_line)
            f.write(new_line)
        f.write("0\n")
        new_goals = []
        for goal in goals:
            tokens = goal.split()
            tokens[2] = str(height - int(tokens[2]) - 1)
            f.write(" ".join(tokens)+"\n")
```

```python
for filename in os.listdir("../mirror/"):
    print(filename)
    mirror("../mirror/"+filename)
```
